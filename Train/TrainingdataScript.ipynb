```python
import os
import pandas as pd

data = []
for i in range(1, 14):
    folder_path = '/content/drive/MyDrive/ProjectData/s' + str(i) + '/txt/'
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, "r") as file:
            text = file.read()
            data.append({'Path': file_path, 'Text': text})

df = pd.DataFrame(data)
print(df)
```

```python
df = df.drop("Path", axis=1)
```

```python
import os
import pandas as pd
from PIL import Image
import numpy as np

data = []

for i in range(1, 14):
    img_folder_path = '/content/drive/MyDrive/ProjectData/s' + str(i) + '/img/'
    for file_name in os.listdir(img_folder_path):
        file_path = os.path.join(img_folder_path, file_name)

        img = Image.open(file_path)

        img_array = np.array(img)

        data.append({'Path': file_path, 'Image': img_array})

df2 = pd.DataFrame(data)


```

```python
df2 = df2.drop("Path", axis=1)
```
